```python
#chanaging the formate of genres i.e into list
import ast
ast.literal_eval('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
def convert(obj):
    L=[]
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L    
        
```

```python
movies['genres']= movies['genres'].apply(convert)
```

```python
movies.head()
```

```python
movies['keywords']=movies['keywords'].apply(convert)
```

```python
#creating a function to get the 3 actors name  
def convert3(obj):
    L=[]
    counter=0
    for i in ast.literal_eval(obj):
        if counter !=3:
            L.append(i['name'])
            counter+=1
        else:
            break
            
    return L    
        
```

```python
movies['cast']=movies['cast'].apply(convert3)
```

```python
movies.head()
```

```python
movies['crew'][0]
```

```python
#make a list of dictory from disnarory 
def fetch_director(obj):
    L=[]
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            
            L.append(i['name'])
        break
    return L    
        
```

```python
movies['crew']=movies['crew'].apply(fetch_director)
```

## In the tags we need to find similarities between them:
problem is :
1.the data is textual data hence mathical operation canot be done

2.even if we count the number of similar words it wont be so effective

hence we do vectorization :
we convert the data i.e these tags i.e movies into vectors

we will be using bag of words technique for vectors:
in this we combine all the tags into one stream of words and find the count of common words in the stream of words


```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors=cv.fit_transform(new_df['tags']).toarray()
```

## to remove the sliently similar name we will use steaming
i.e ['loved','loving','love'] changed to ['love','love','love']
 we will do this one the new_df data directly


```python
ps.stem('loving')
```

```python
#library will has stemmer nltk to be imperted
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps= PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)    
```

```python
new_df['tags']=new_df['tags'].apply(stem)
```

# find the similarities between the vectors :
since its a higher dimensional space i.e 4056*5000 , eculdian distance won't be applicable instead we will use cosine distance:


```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cosine_similarity(vectors).shape
```

```python
similarity=cosine_similarity(vectors)
```

```python
similarity[0]
```

## since above given values are similarity score of 0 index tag with all the other tags, if we sort them their similarity index will be lost
hence we will assign them to be fixed with their index by converting them into tuples by using 
## enumerate 
as sorted will so it in ascending order we will use reverse to get in reverse order coz the we need most similar list


```python
sorted(list(enumerate(similarity[0])),reverse=True, key=lambda x:x[1] )[1:6]
```

```python
def recommend(movie):
#finding the index of movie    
    
    movie_index=new_df[new_df['title']==movie].index[0]
    distances=similarity[movie_index]
    movies_list=sorted(list(enumerate(distances)),reverse=True, key=lambda x:x[1] )[1:6]
    for i in movies_list:
        print(new_df.iloc[i[0]].title)

    
```

```python
recommend('Avatar')
```
